## 1. My lane as an ML task (type)

My chosen lane is CTR / Engagement Opportunity Scoring.

Task type: Scoring.

This is a scoring problem because the goal is to assign each page an opportunity score based on signals such as visibility, clicks, CTR, and engagement. The model will rank pages from highest to lowest opportunity so that content teams can prioritize which pages need optimization.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## 2. Target or Proxy

The target is an engagement opportunity score based on observed page performance metrics.

The target will be created using available signals such as impressions, clicks, CTR, and engagement metrics. Pages with high visibility (impressions) but lower-than-expected CTR will be considered higher opportunities for improvement.

The label is a defined proxy because "content opportunity" is not directly measured in the dataset. Instead, we create a practical proxy from observed user engagement behavior to identify pages that may benefit from optimization.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## 3. Success Metric

The success metric for this task is Ranking Quality, measured using NDCG (Normalized Discounted Cumulative Gain).

A higher NDCG score means the model is better at placing high-opportunity pages near the top of the recommendation list. A good result would be a model that consistently ranks pages with the greatest improvement potential above lower-priority pages.

This metric is suitable because the final goal is not only to predict a value, but to create a useful ranked list of pages for content optimization.

## 4. The Unit of Analysis

The unit of analysis is one content page.

Each row represents one page and contains its search performance information, such as impressions, clicks, CTR, and engagement signals. The model will use these page-level features to calculate an opportunity score and identify pages that may benefit from optimization.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## 5. Why ML Beats a Fixed Rule Here

A fixed rule cannot capture the complexity of content performance because many factors influence CTR and engagement, such as impressions, position, content type, topic, and user behavior.

For example, a page with a low CTR may not always be a problem if it has low visibility or serves a different purpose. ML can learn patterns from multiple signals together and create more accurate opportunity scores than a simple if-statement.

The model can help prioritize pages by considering multiple factors instead of relying on one manually chosen threshold.
